```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline
```

```python
# Create DF
my_df = pd.DataFrame(randn(25, 4), columns=['Mon', 'Tues', 'Wed', 'Thurs'])
my_df
```

```python
# Aera plat
my_df.abs().plot(kind='area')
```

```python
# Shading (transparence)
my_df.abs().plot(kind='area', alpha=0.3)
```

```python
# Title
my_df.abs().plot(kind='area', alpha=0.3, title="My awesome aera plot!")
```

```python
# Legend
my_df.abs().plot(kind='area', alpha=0.3, title="My awesome aera plot!", legend=False)
```

```python
# Plot only one column
my_df['Mon'].abs().plot(
    kind='area', alpha=0.3, title="My awesome aera plot!", legend=False)
```

```python
# Plot 2 columns
my_df[['Mon', 'Thurs']].abs().plot(
    kind='area', alpha=0.3, title="My awesome aera plot!", legend=False)
```

```python
# Add table
my_df[['Mon', 'Thurs']].abs().plot(
    kind='area', alpha=0.3, title="My awesome aera plot!", legend=False,
    table=True)
```
