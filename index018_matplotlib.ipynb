```python
%matplotlib inline
import pandas as pd
import numpy as np
from numpy.random import rand
```

```python
# Create a quick DF
my_df = pd.DataFrame(rand(100, 4), columns=['Mon', 'Tues', 'Wed', 'Thur'])
my_df
```

```python
# Histogram
my_df['Wed'].hist()
```

```python
# Histogram with Smaller bins, 50, 20
# Histogram
my_df['Wed'].hist(bins=50)
```

```python
# Another way without grid bars
my_df['Wed'].plot(kind='hist', bins=20)
```

```python
my_df['Wed'].plot(kind='hist', bins=20, grid=True, legend=True)
```
