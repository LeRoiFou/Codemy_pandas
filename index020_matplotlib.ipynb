```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline
```

```python
# Create DF
my_df = pd.DataFrame(randn(25, 4), columns=['Mon', 'Tues', 'Wed', 'Thurs'])
my_df.head()
```

```python
# Bar plat
my_df.abs().plot(kind='bar')
```

```python
# Bar plat stacked
my_df.abs().plot(kind='bar', stacked=True)
```

```python
# Shading
my_df.abs().plot(kind='bar', stacked=True, alpha=0.4)
```

```python
# Title
my_df.abs().plot(kind='bar', stacked=True, title="My awesome bar chart!")
```

```python
# Legend
my_df.abs().plot(kind='bar', stacked=True, title="My awesome bar chart!",
                 legend=False)
```

```python
# Second way
my_df.abs().plot.bar()
```

```python
# Second way stacked
my_df.abs().plot.bar(stacked=True)
```

```python
# Second way shading
my_df.abs().plot.bar(stacked=True, alpha=0.4)
```

```python
# Second way title
my_df.abs().plot.bar(stacked=True, title="My awesome second way chart graph!")
```

```python
# Second way legend
my_df.abs().plot.bar(stacked=True, title="My awesome second way chart graph!",
                     legend=False)
```

```python
# Grid lines
my_df.abs().plot.bar(stacked=True, title="My awesome second way chart graph!",
                     legend=False, grid=True)
```
