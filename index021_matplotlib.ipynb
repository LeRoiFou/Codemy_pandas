```python
import pandas as pd
import numpy as np
from numpy.random import randn
# %matplotlib inline
```

```python
# Create DF
my_df = pd.DataFrame(randn(25, 4), columns=['Mon', 'Tues', 'Wed', 'Thurs'])
my_df.head()
```

```python
# Line plot
my_df.plot(kind='line')
```

```python
# Line plot absolute
my_df.abs().plot(kind='line')
```

```python
# Line width
my_df.abs().plot(kind='line', lw=5)
```

```python
# Line single
my_df['Mon'].abs().plot(kind='line')
```

```python
# Line more than one
my_df[['Mon', 'Tues']].abs().plot(kind='line')
```

```python
# Size
my_df[['Mon', 'Tues']].abs().plot(kind='line', figsize=(10, 5))
```

```python
# Title
my_df[['Mon', 'Tues']].abs().plot(kind='line', figsize=(10, 5),
                                  title="My awesome line chart!")
```

```python
# Shading
my_df[['Mon', 'Tues']].abs().plot(kind='line', figsize=(10, 5),
                                  alpha=0.3)
```

```python
# Legend
my_df[['Mon', 'Tues']].abs().plot(kind='line', figsize=(10, 5),
                                  legend=False)
```

```python
# Second way chart
my_df.plot.line()
```
