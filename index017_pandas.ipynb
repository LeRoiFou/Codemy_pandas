```python
import pandas as pd
import numpy as np
```

```python
# Create data
stuff = {
    "Corporation":['Apple', 'Google', 'Meta', 'Apple', 'Google', 'Meta'],
    "Employees":['John', 'April', 'Wes', 'Beth', 'Justin', 'Steph'],
    "Salary":[200, 220, 190, 130, 120, 150]}
```

```python
# Create DF
my_df = pd.DataFrame(stuff)
my_df
```

```python
# Sort salaries from lowest to highest
my_df.sort_values(by='Salary')
```

```python
# Sort salaries from highest to lowest
my_df.sort_values(by='Salary', ascending=False)
```

```python
# Not permanent without inplace=True
my_df
```

```python
# Inplace
my_df.sort_values(
    by='Salary', ascending=False, inplace=True)
my_df
```

```python
# Sort String Alphabetically
my_df.sort_values(by='Corporation')
```

```python
# Sort String Alphabetically descending
my_df.sort_values(by='Corporation', ascending=False)
```
