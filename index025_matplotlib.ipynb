```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline
```

```python
# Create DF
my_df = pd.DataFrame(randn(500, 4), columns=['Mon', 'Tues', 'Wed', 'Thurs'])
my_df.head()
```

```python
# %pip install scipy
```

```python
# Density and kernel density estimation plots with scipy
my_df.plot(kind='kde')
```

```python
# KDE second way to learn shift tab
my_df.plot.kde()
```

```python
# Density plot
my_df.plot.density()
```

```python
# Other method
my_df.plot(kind='density')
```

```python
# Change stuff
my_df.plot.kde(alpha=0.4, lw=4)
```

```python
# Grid
my_df.plot.kde(alpha=0.4, lw=4, grid=True)
```

```python
# Title
my_df.plot.kde(alpha=0.4, lw=4, grid=True, title="My Awesome KDE Plot!")
```
