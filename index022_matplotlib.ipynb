```python
import pandas as pd
import numpy as np
from numpy.random import randn
# %matplotlib inline
```

```python
# Create DF
my_df = pd.DataFrame(randn(500, 4), columns=['Mon', 'Tues', 'Wed', 'Thurs'])
my_df.head()
```

```python
# Scatter plot
my_df.plot(kind='scatter', x='Mon', y='Tues', c='Wed')
```

```python
# Colors : https://matplotlib.org/stable/tutorials/colors/colormaps.html
my_df.plot(kind='scatter', x='Mon', y='Tues', c='Wed', cmap='magma')
```

```python
# By size
my_df.plot(kind='scatter', x='Mon', y='Tues', s=my_df['Wed']*50, cmap='magma')
```

```python
# By size easier to read with alpha
my_df.plot(kind='scatter', x='Mon', y='Tues', 
           s=my_df['Wed']*200, cmap='magma', alpha=0.4)
```

```python
# Title etc...
my_df.plot(kind='scatter', x='Mon', y='Tues', 
           s=my_df['Wed']*200, cmap='magma', alpha=0.4,
           title="My awesom ScatterPlot!")
```
