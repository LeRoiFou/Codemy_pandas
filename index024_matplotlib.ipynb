```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline
```

```python
# Create DF
my_df = pd.DataFrame(randn(500, 4), columns=['Mon', 'Tues', 'Wed', 'Thurs'])
my_df.head()
```

```python
# Hexplots
my_df.plot(kind='hexbin', x='Mon', y='Tues')
```

```python
# hexplots gridsize
my_df.plot(kind='hexbin', x='Mon', y='Tues', gridsize=10)
```

```python
# hexplots color
my_df.plot(kind='hexbin', x='Mon', y='Tues', gridsize=10, cmap='Spectral')
```

```python
# Other method
my_df.plot.hexbin(x='Mon', y='Tues', gridsize=10)
```

```python
# Title
my_df.plot.hexbin(x='Mon', y='Tues', gridsize=10,
           title='My Awesome Hex Plot!')
```

```python
# Grid
my_df.plot.hexbin(x='Mon', y='Tues', gridsize=10,
           title='My Awesome Hex Plot!')
```
